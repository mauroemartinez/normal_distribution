# Plot Normal Distribution with given Mean and Standard Deviation

### Source: https://www.youtube.com/watch?v=Xm5QnHGw0bc
### Help from: https://chat.openai.com/chat

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mean = int(input("Mean = "))
std = int(input("Standard Deviation = "))
```

```python
# create linespace with 1000 numbers, 4*std to the left and to the right
x = np.linspace(mean-4*std, mean+4*std, 1000)

# plot normal distribution, loc is mean and scale is standard deviation
# for each value in X we calculate the value in the norm distribution
# said normal distribution is technically known as Probability Density Function
y = norm(loc=mean, scale=std).pdf(x)

plt.figure(figsize=(10,5))
plt.plot(x, y, c="b", linewidth=3)

#set y axis to start from 0
plt.ylim(ymin=0)

# plot a vertical line right on the mean with linestyle and label (legend)
plt.axvline(
    mean,
    ls="-.",
    c="silver",
    label=f"norm dist with μ={mean}, σ={std}"
)
# same with 2 standard deviations to the left and to the right
plt.axvline(
    mean-std,
    ls="-.",
    c="orange",
    label=f"mean-std"
)
plt.axvline(
    mean+std,
    ls="-.",
    c="orange",
    label=f"mean + std"
)
# between 2 standard deviations to the left and to the right we get approx 95% of probability 
plt.axvline(
    mean-2*std,
    ls="-.",
    c="r",
    label=f"mean - 2std"
)
plt.axvline(
    mean+2*std,
    ls="-.",
    c="r",
    label=f"mean + 2std"
)

# fill areas with color
plt.fill_between(
    x, y,
    where=np.logical_and(x >= mean-2*std,x < mean-std),
    color='lightgreen'
)
plt.fill_between(
    x, y,
    where=np.logical_and(x >= mean-std, x <= mean+std),
    color='green',
    alpha=0.7
)
plt.fill_between(
    x, y,
    where=np.logical_and(x > mean+std, x <= mean+2*std),
    color='lightgreen'
)
plt.fill_between(
    x, y,
    where=x < mean-2*std,
    color='r'
)
plt.fill_between(
    x, y,
    where=x > mean+2*std,
    color='r'
)

# add annotations with probability by area
plt.text(
    (((mean-std)+(mean-2*std)))/2.0033,
    0.01,
    "13.6%",
    fontsize=12,
    color="black",
    fontweight="bold"
)
plt.text(
    (((mean+std)+(mean+2*std)))/2.0033,
    0.01,
    "13.6%",
    fontsize=12,
    color="black",
    fontweight="bold"
)
plt.text(
    (mean-std+mean)/2.0033,
    0.03,
    "34.1%",
    fontsize=12,
    color="black",
    fontweight="bold"

)
plt.text(
    (mean+std+mean)/2.0033,
    0.03,
    "34.1%",
    fontsize=12,
    color="black",
    fontweight="bold"
)

# show legend and show plot
plt.legend()
plt.show()
```
